# Birthday

## Birthdate 타입 변경

```python
import pandas as pd
csv_data = open('DataCleansing.csv','r')
df = pd.read_csv(csv_data, header=0, index_col=None)
```

```python
df.head()
```

```python
import datetime as dt
df['Birthdate'] = df['Birthdate'].apply(lambda x: dt.datetime.strptime(x, '%m.%d.%y'))
df.info()
# to do
```

```python
df.head()
```

## 잘못된 날짜 조정

replace 함수 활용

```python
df.loc[0, 'Birthdate'].year
```

```python
for index in df.index:
    elem = df.loc[index,'Birthdate']
    if(elem.year > 2017):
        df.loc[index, 'Birthdate'] = df.loc[index, 'Birthdate'].replace(year=elem.year-100)
        
```

```python
df.head()
```

```python
df['Birthdate'] = df['Birthdate'].apply(lambda x: x.replace(year = x.year-100) if (x.year>2017) else x)
```

```python
df
```

apply와 replace 함께 활용

## 나이 조정

```python
df.sort_values('Birthdate')  #ascending or descending sort
```

생년월일 이상치 확인

```python
Birthdate = df.Birthdate.value_counts()
Birthdate[Birthdate > 1]
```

나이 이상치 확인

나이 계산 다시

```python
from datetime import date, timedelta
```

```python
#df['Age'] = df['Birthdate'].apply(lambda x: int((date.today() - x.date())/ timedelta(days = 365))) #error
df['Age'] = (date.today() - df['Birthdate']) / timedelta(days = 365)
```
