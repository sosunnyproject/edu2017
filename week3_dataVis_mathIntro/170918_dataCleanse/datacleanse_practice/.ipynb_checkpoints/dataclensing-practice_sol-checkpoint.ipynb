# Birthday

## Birthdate 타입 변경

```python
import pandas as pd
csv_data = open('DataCleansing.csv','r')
df = pd.read_csv(csv_data, header=0, index_col=None)
```

```python
import datetime as dt
df['Birthdate'] = df['Birthdate'].apply(lambda x: dt.datetime.strptime(x,'%m.%d.%y'))
df.info()
```

```python
df
```

## 잘못된 날짜 조정

```python
for index in df.index:
    elem = df.ix[index, 'Birthdate']
    if (elem.year>2017):
        df.ix[index, 'Birthdate'] = df.ix[index, 'Birthdate'].replace(year=elem.year-100)
    '''
    df.ix[index, 'Birthdate'] = df.ix[index, 'Birthdate'].replace(
        year=df.ix[index, 'Birthdate'].year-100 if df.ix[index, 'Birthdate'].year>2017 
        else df.ix[index, 'Birthdate'].year)
    '''
```

```python
df['Birthdate'] = df['Birthdate'].apply(lambda x: x.replace(year=x.year-100) if (x.year>2017) else x)
```

```python
df
```

## 나이 조정

```python
df.sort_values('Birthdate')
```

생년월일 이상치 확인

```python
Birthdate = df.Birthdate.value_counts()
Birthdate[Birthdate > 1]
```

나이 이상치 확인

```python
df[df['Age']<0]
```

나이 계산 다시

```python
from datetime import date, timedelta
df['Age'] = df['Birthdate'].apply(lambda x: int((date.today()-x.date())/timedelta(days=365)))
# df['Age']= (date.today() - df['Birthdate']) / timedelta(days=365)
```

# Region

지역 종류 파악

```python
df_R = df.copy()
```

```python
df_R.Region.value_counts()
```

같은 지역인데 이름이 다른 경우 조정

```python
df_R['Region'] = df_R['Region'].apply(lambda x: 'South' if x == ' South' else 'South')
df_R['Region'] = df_R['Region'].apply(lambda x: 'East' if x == ' East' else 'East')
df_R['Region'] = df_R['Region'].apply(lambda x: 'West' if x == ' West' else 'West')
df_R['Region'] = df_R['Region'].apply(lambda x: 'North' if x == ' North' else 'North')
```

```python
df_R = df_R.replace({'Region': {' South': 'South', ' East': 'East', ' West': 'West', ' North': 'North'}})
```

```python
df_R.Region = df_R.Region.str.strip()
```

```python
df_R.Region.value_counts()
```
