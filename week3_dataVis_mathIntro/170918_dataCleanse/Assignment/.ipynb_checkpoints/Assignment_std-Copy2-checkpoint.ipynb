### Q16) 년도별 'lesl'로 시작하는 이름을 가진 아이들 중에 여자 아이의 비율, 남자 아이의 비율 구하기

```python
def leslprop(df):
    total = df.F +df.M
    f = df.F / total
    m = df.M / total
    return pd.Series({'F':f, 'M':m})
tlCopy = totalLesl.copy()
tlCopy[['F', 'M']] = tlCopy.apply(leslprop, axis = 1) 
tlCopy
```
