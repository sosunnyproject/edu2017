### 익명 함수를 이용한 정규분포함수의 시각화

```python
mu=0
sigma = 1
f = lambda x: 1/np.sqrt(2*np.pi*sigma)*np.exp(-0.5*(x-mu)**2/sigma)
```

```python
x = np.linspace(-5,5, 10000)
plt.plot(x,f(x), linewidth = 5)
plt.xlim(-5,5)
plt.show()
```

### 데이터 샘플링

```python
from scipy.stats import norm

plt.hist(norm.rvs(size=10000), bins = 80, color='orange', normed = True)
plt.plot(x,f(x), color='red', linewidth = 5)
plt.xlim(-5,5)
plt.show()
```

### 평균과 분산의 변화에 따른 정규분포의 확률 밀도 함수의 비교

```python
from scipy.stats import norm
x = np.linspace(-7,7, 10000)
mu = [0,0,0,-2]
sigma = [0.5,1,2,1]
for i in range(4):
    plt.plot(x, norm.pdf(x,mu[i],sigma[i]), linewidth = 5, label=r'$\mu = {}, \sigma = {}$'.format(mu[i],sigma[i]))
plt.plot([-7,7],[0,0],'k')
plt.xlim(-7,7)

plt.legend(loc='upper right')
plt.show()
```

### 평균과 분산의 변화에 따른 정규분포의 누적 분포 곡선의 비교

```python
x = np.linspace(-5,5, 10000)
mu = [0,0,0,-2]
sigma = [0.5,1,2,1]
for i in range(4):
    plt.plot(x, norm.cdf(x,mu[i],sigma[i]), linewidth=5, label=r'$\mu = {}, \sigma = {}$'.format(mu[i],sigma[i]))
plt.xlim(-5,5)
plt.ylim(-0.02,1.05)
plt.plot([-5,5],[0,0],'k')
plt.plot([-5,5],[1,1],'k--')
plt.legend(loc='lower right')
plt.show()
```

### 자유도(Degree of freedom)의 변화에 따른 χ2 분포의 확률 밀도 함수의 비교

```python
from scipy.stats import chi2
x = np.linspace(0,8, 10000)

for i in range(1,6):
    plt.plot(x, chi2.pdf(x,i), linewidth=5, label=r'$\nu = {}$'.format(i))
plt.plot([0,8],[0,0],'k')
plt.xlim(0,8)
plt.ylim(-0.02, 0.6)

plt.legend(loc='upper right')
plt.show()
```

### 자유도(Degree of freedom)의 변화에 따른 χ2 분포의 누적 밀도 함수의 비교

```python
for i in range(1,6):
    plt.plot(x, chi2.cdf(x,i), linewidth=5, label=r'$\nu = {}$'.format(i))
plt.plot([0,8],[0,0],'k')
plt.xlim(0,8)
plt.ylim(-0.02, 1.02)

plt.legend(loc='lower right')
plt.show()
```

### 자유도(Degree of freedom)의 변화에 따른 t-분포의 확률 밀도 함수의 비교

```python
from scipy.stats import t
x = np.linspace(-5,5, 10000)

df = [1,2,3,4,30]

for nu in df:
    plt.plot(x, t.pdf(x,nu), linewidth=3, label=r'$\nu = {}$'.format(nu))
from scipy.stats import norm
plt.plot(x, norm.pdf(x),linewidth=3, linestyle='--', label='Normal')
plt.plot([-5,5],[0,0],'k')
plt.xlim(-5,5)
#plt.ylim(-0.02, 0.6)

plt.legend(loc='upper right')
plt.show()
```

### 자유도(Degree of freedom)의 변화에 따른 t-분포의 누적 밀도 함수의 비교

```python
from scipy.stats import t
x = np.linspace(-5,5, 10000)

df = [1,2,3,4,30]

for nu in df:
    plt.plot(x, t.cdf(x,nu), linewidth=3, label=r'$\nu = {}$'.format(nu))
from scipy.stats import norm
plt.plot(x, norm.cdf(x), linewidth=3, linestyle = '--', label='Normal')
plt.plot([-5,5],[0,0],'k')
plt.xlim(-5,5)

plt.legend(loc='lower right')
plt.show()
```

### Law of Large Number

```python
from scipy.stats import norm
import seaborn as sns
mu=[]
mu1=[]
n1=20
n2=200
for i in range(10000):
    x = norm.rvs(size=n1)
    mu.append(np.mean(x))
    x = norm.rvs(size=n2)
    mu1.append(np.mean(x))
```

```python
plt.figure(figsize=(7,10))
sns.set_style("ticks")
plt.hist(norm.rvs(size=10000), bins = 160, label='Original', normed = True)
x=np.linspace(-2,2, 1000)
plt.plot(x, norm.pdf(x),linewidth=5, color='blue', linestyle='--')
plt.hist(mu, bins=80, label=r'$n={}$'.format(n1), normed=True )
sns.kdeplot(np.array(mu),linewidth=5,color='green', bw=0.05)
plt.hist(mu1, bins=160, label=r'$n={}$'.format(n2), normed=True )
sns.kdeplot(np.array(mu1),linewidth=5,color='purple', bw=0.05)
plt.xlim(-2,2)
plt.ylim(-0.02,4)
#plt.plot([-5,5],[0,0],'k')
#plt.plot([-5,5],[1,1],'k--')
plt.legend(loc='upper right')
plt.show()
```

### Central Limit Theorem

```python
from scipy.stats import chi2

x = np.linspace(0,8,1000)
plt.figure(figsize=(7,4))

plt.plot(x, chi2.pdf(x,3), linewidth=5, label='Original')
x = np.linspace(-3,3,1000)
plt.plot(x, norm.pdf(x), color='red', linewidth=5, label='normal')

y=[]
number = 5
for i in range(10000):
    x = chi2.rvs(3,size=number)
    y.append(np.sqrt(number/6)*(np.mean(x) - 3))

plt.hist(y, bins=80, label=r'$n = {}$'.format(number), color='orange', normed=True )
plt.xlim(-3,8)
plt.ylim(-0.02,0.5)
plt.legend(loc='upper right')
plt.show()
```
